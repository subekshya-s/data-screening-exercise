## Analyzing

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    "Facility": [
        "ADAMS COUNTY DET CENTER",
        "SOUTH TEXAS ICE PROCESSING CENTER",
        "STEWART DETENTION CENTER",
        "WINN CORRECTIONAL CENTER",
        "OTAY MESA DETENTION CENTER",
        "ELOY FEDERAL CONTRACT FACILITY",
        "MOSHANNON VALLEY PROCESSING CENTER",
        "MONTGOMERY ICE PROCESSING CENTER",
        "CENTRAL LOUISIANA ICE PROCESSING CENTER",
        "DENVER CONTRACT DETENTION FACILITY"
    ],
    "Level A": [1876.46, 1299.08, 743.31, 1013.02, 800.83, 1041.89, 205.25, 149.56, 410.97, 624.57],
    "Level B": [266.43, 106.13, 194.93, 214.77, 470.12, 45.54, 96.47, 561.43, 221.21, 190.51],
    "Level C": [6.72, 188.72, 371.56, 187.01, 45.48, 114.69, 542.28, 355.27, 344.79, 189.66],
    "Level D": [4.26, 68.20, 328.04, 94.58, 46.22, 125.61, 402.06, 160.98, 207.16, 141.37]
}

df = pd.DataFrame(data)

# Sort by total population
df["Total"] = df[["Level A","Level B","Level C","Level D"]].sum(axis=1)
df = df.sort_values("Total", ascending=False)

# Plot stacked bars
plt.figure(figsize=(13, 8))

plt.barh(df["Facility"], df["Level A"], label="Level A")
plt.barh(df["Facility"], df["Level B"], left=df["Level A"], label="Level B")
plt.barh(df["Facility"], df["Level C"], left=df["Level A"]+df["Level B"], label="Level C")
plt.barh(df["Facility"], df["Level D"], left=df["Level A"]+df["Level B"]+df["Level C"], label="Level D")

plt.xlabel("Population")
plt.title("Top 10 Detention Centers in the US by Custody Level Distribution")
plt.legend()
plt.tight_layout()
plt.show()
```
